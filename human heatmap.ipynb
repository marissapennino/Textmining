```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('/Users/marissapennino1/Desktop/New-Human-cardiac-tissue-specific-proteome/results/caseolap.csv')
```

```python
list_proteins = df['protein'].unique()
y = df['protein']
df1 = df.drop(['protein'], axis=1)
column_name_list = df1.columns.tolist()
plt.figure(figsize=(10, 10))
fig3=sns.heatmap(df1, xticklabels=column_name_list , yticklabels=False, cmap="YlGnBu")
plt.title("Human Heart Tissue CASEOLAP", fontsize =20)
plt.savefig('fig3.jpg',dpi=300)
```

```python
df_filtered=df.loc[~(df.select_dtypes(float) < 0.1).all(axis="columns")]


list_proteins = df_filtered['protein'].unique()
y = df_filtered['protein']
df1 = df_filtered.drop(['protein'], axis=1)
column_name_list = df1.columns.tolist()
plt.figure(figsize=(10, 12))
fig4a=sns.heatmap(df1, xticklabels=column_name_list , yticklabels=list_proteins, cmap="YlGnBu", fmt='.2f', annot=True)
plt.title("Human Heart Tissue CASEOLAP", fontsize =20)
plt.savefig('fig4a.jpg',dpi=300)
```

```python
df_filtered=df.loc[~(df.select_dtypes(float) < 0.1).all(axis="columns")]


list_proteins = df_filtered['protein'].unique()
y = df_filtered['protein']
df1 = df_filtered.drop(['protein'], axis=1)
column_name_list = df1.columns.tolist()
plt.figure(figsize=(10, 12))
fig4b=sns.heatmap(df1, xticklabels=column_name_list , yticklabels=list_proteins, cmap="YlOrRd", fmt='.2f', annot=True)
plt.title("Human Heart Tissue CASEOLAP", fontsize =20)
plt.savefig('fig4b.jpg',dpi=300)
```

```python
fig5a=sns.kdeplot(df_filtered.IHD, df_filtered.CM, cmap="Reds", shade=True, shade_lowest=True)
plt.title("IHD, CM CASEOLAP Scores", fontsize =20)
plt.savefig('fig5a.jpg',dpi=150)
```

```python
fig5b=sns.kdeplot(df_filtered.VD, df_filtered.VOO, cmap="Reds", shade=True, shade_lowest=True)
plt.title("VD, VOO CASEOLAP Scores", fontsize =20)
plt.savefig('fig5b.jpg',dpi=150)
```
